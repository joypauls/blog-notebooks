```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use("./blog.mplstyle")
```

```python

labels = ["Renewables", "Nuclear", "Natural Gas", "Coal"]
sizes = [33, 21, 24, 22]
colors = ["#57ae5b", "#98e19b", "#C4BAC7", "#918295"]

carbon_free_sizes = [54, 46, 0, 0]
carbon_free_colors = ["#2D8444", "none", "none", "none"]

fig, ax = plt.subplots(figsize=(8, 8))

wedges, _ = ax.pie(
    sizes,
    startangle=90,
    colors=colors,
    wedgeprops=dict(width=0.25, edgecolor="white"),
    radius=0.9
)

inner_wedges, _ = ax.pie(
    carbon_free_sizes,
    startangle=90,
    colors=carbon_free_colors,
    wedgeprops=dict(width=0.07, edgecolor="white"),
    radius=0.65
)

# Add labels next to each wedge
for i, (wedge, label, size) in enumerate(zip(wedges, labels, sizes)):
    angle = (wedge.theta2 + wedge.theta1) / 2
    factor = 1.08 if i in [1, 3] else 1.15
    x = np.cos(np.deg2rad(angle)) * factor
    y = np.sin(np.deg2rad(angle)) * factor
    ha = "left" if x > 0 else "right"

    ax.text(
        x, y,
        f"{label}\n{size}%",
        # ha=ha,
        ha="center",
        va="center",
        fontsize=12,
        bbox=dict(boxstyle="square,pad=0.4", fc="white", ec="gray", lw=0.6)
    )

# add padding to annotation
ax.text(0, 0, f"{54}%", ha="center", va="center", fontsize=28, weight="bold")
# write text below
ax.text(0, -0.11, f"Carbon-Free", ha="center", va="center", fontsize=16, weight="regular")

ax.set_title("MN Electricity Generation Mix (2023)", fontsize=18)
ax.text(0, 1.25, "Data Source: U.S. EIA", ha="center", va="center", fontsize=12)
ax

# add padding
# fig.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1)


# plt.subplots_adjust(left=0.2, right=0.8, top=0.9, bottom=0.1)
plt.show()


```
