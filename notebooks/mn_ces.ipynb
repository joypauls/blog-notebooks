```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use("./blog.mplstyle")
```

```python


# Data
labels = ['Renewables', 'Nuclear', 'Natural Gas', 'Coal']
sizes = [33, 21, 24, 22]
colors = ['#66bb6a', '#42a5f5', '#ffa726', '#b0bec5']

# Carbon-free indicator (inner ring)
carbon_free_sizes = [54, 46, 0, 0]
carbon_free_colors = ['#388e3c', 'none', 'none', 'none']  # darker tones to match outer

fig, ax = plt.subplots(figsize=(8, 8))

# Outer ring — main electricity mix
wedges, _ = ax.pie(
    sizes,
    startangle=90,
    colors=colors,
    wedgeprops=dict(width=0.3, edgecolor='white'),
    radius=1.0
)

# Inner ring — carbon-free highlight
inner_wedges, _ = ax.pie(
    carbon_free_sizes,
    startangle=90,
    colors=carbon_free_colors,
    wedgeprops=dict(width=0.07, edgecolor='white'),
    radius=0.7
)

# Add labels next to each wedge
for i, (wedge, label, size) in enumerate(zip(wedges, labels, sizes)):
    angle = (wedge.theta2 + wedge.theta1) / 2
    x = np.cos(np.deg2rad(angle)) * 1.2
    y = np.sin(np.deg2rad(angle)) * 1.2
    ha = 'left' if x > 0 else 'right'

    ax.text(
        x, y,
        f"{label}\n{size}%",
        ha=ha,
        va='center',
        fontsize=12,
        bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="gray", lw=0.6)
    )

# Center carbon-free label
ax.text(0, 0, f'{54}%\nCarbon-Free', ha='center', va='center', fontsize=18, weight='bold')

# Title
ax.set_title('Minnesota Electricity Generation Mix — 2023', fontsize=18, pad=20)
plt.tight_layout()
plt.show()


```
