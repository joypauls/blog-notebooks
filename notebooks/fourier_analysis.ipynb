```python
import numpy as np
import matplotlib.pyplot as plt


# Function to compute the Fourier series approximation of a square wave
def square_wave_fourier_series(x, n):
    sum = np.zeros_like(x)
    for k in range(1, n+1):
        n = 2 * k - 1  # Only odd harmonics
        sum += (1 / n) * np.sin(n * x)
    sum = (4 / np.pi) * sum
    return sum

# Define the domain
x = np.linspace(-np.pi, np.pi, 1000)

# Original square wave function
def square_wave(x):
    return np.sign(np.sin(x))

# Number of Fourier series terms
n = 5  # Change N to see the effect of more terms

# Compute the Fourier series approximation
f_approx = square_wave_fourier_series(x, n)


plt.style.use("seaborn-v0_8-whitegrid")
plt.figure(figsize=(10, 6))
plt.plot(x, square_wave(x), label="Square Wave", linewidth=2)
plt.plot(x, f_approx, label=f"Fourier Series Approximation (n={n})")
plt.title("Fourier Series Approximation of a Square Wave")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
# plt.grid()
plt.show()
```

```python
n = 5
x = np.linspace(-np.pi, np.pi, 1000)

plt.style.use("seaborn-v0_8-whitegrid")
plt.figure(figsize=(10, 6))
for k in range(1, n + 1):
    n = 2 * k - 1  # Only odd harmonics
    sinusoid = (1 / n) * np.sin(n * x)
    plt.plot(x, sinusoid, label=f"Fourier Series Term {k} (n={n})")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
# plt.grid()
plt.show()
```
