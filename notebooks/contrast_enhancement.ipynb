```python
# gray = cv2.cvtColor(cv2.imread("../data/hist_eq/PA200757.png"), cv2.COLOR_BGR2GRAY)
color = cv2.imread("../data/hist_eq/PA200757.jpg")


# print(gray.shape)
print(color.shape)
```

```python
def plot(img, title, color = True):
    # plt.figure(figsize=(10,10))
    plt.style.use("seaborn-v0_8-white")
    plt.figure(figsize=(8,5))
    plt.title(title)
    if color:
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    else:
        plt.imshow(img, cmap="gray")
    plt.show()
```

```python
def enhance_contrast(img, alpha, beta):
    return cv2.convertScaleAbs(img, alpha=alpha, beta=beta)

# def enhance_contrast_color(img, alpha, beta):
#     img_copy = img.copy()
#     # cv2.convertScaleAbs(img, alpha=alpha, beta=beta)
#     return cv2.convertScaleAbs(img, alpha=alpha, beta=beta)
```

```python
plt.style.use("seaborn-v0_8-whitegrid")

plt.hist(cv2.cvtColor(color, cv2.COLOR_BGR2RGB).ravel(), 256, [0, 256], alpha=0.7)
plt.show()


plt.hist(cv2.cvtColor(enhance_contrast(color, 1.7, -80), cv2.COLOR_BGR2RGB).ravel(), 256, [0, 256], alpha=0.7)
plt.show()
```

```python
h = cv2.cvtColor(color, cv2.COLOR_BGR2RGB).ravel()

plt.style.use("seaborn-v0_8-whitegrid")
plt.figure(figsize=(8,5))

plt.hist(h, 256, [0, 256], alpha=0.7)
plt.axvspan(0, h.min(), color="red", alpha=0.3, lw=0)
plt.axvspan(h.max(), 255, color="red", alpha=0.3, lw=0)
plt.title("Original Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Count")
plt.show()
```
